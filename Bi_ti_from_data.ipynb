```r
head_data <- head(data, 20)
```

```r
data_bigrams <- head_data %>% unnest_tokens(bigram, text, token = "ngrams", n = 2)
head(data_bigrams)
```

```r
bi_sep <- data_bigrams  %>% separate(bigram, c('word1','word2'), sep = ' ')
head(bi_sep)
```

```r
bigrams_filtered <- bi_sep %>%
  filter(!word1 %in% stop_words$word) %>%
  filter(!word2 %in% stop_words$word)
```

```r
bigrams_united <- bigrams_filtered  %>% unite(bigrams, word1, word2, sep = ' ')
head(bigrams_united)
```

```r
bigrams_united  %>% group_by(trn_id)  %>% summarise(bigrams=paste(bigrams, collapse=","))
```

### Bigrams & Trigrams

```r
data %>%
  unnest_tokens(bigram, text, token = "ngrams", n = 2) %>%
  separate(bigram, c("word1", "word2"), sep = " ") %>%
  filter(!word1 %in% stop_words$word,
         !word2 %in% stop_words$word) %>% unite(bigram, word1, word2, sep = ' ')  %>% group_by(trn_id)  %>% summarise(bigram=paste(bigram, collapse=",")) -> bigrams_united
```

```r
head_data %>%
  unnest_tokens(trigram, text, token = "ngrams", n = 3) %>%
  separate(trigram, c("word1", "word2", "word3"), sep = " ") %>%
  filter(!word1 %in% stop_words$word,
         !word2 %in% stop_words$word,
         !word3 %in% stop_words$word) %>% unite(trigram, word1, word2, word3, sep = ' ')  %>% group_by(trn_id)  %>% summarise(trigram=paste(trigram, collapse=",")) -> trigrams_united
```
